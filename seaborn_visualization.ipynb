# Seaborn Visualization Cheat Sheet

[Seaborn](https://seaborn.pydata.org/) is a Python data visualization library based on matplotlib. It works very weel with Pandas DataFrame

This cheat sheet, is a collection of the most commont type of plot that can be useful during the Explanatory Data Analysis **EDA** phase



## Common Import

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

## Dataset load for testing

```python
ds = sns.load_dataset('tips')
```

## Show and Set the style

```python
sns.axes_style()
```

```python
sns.set_style('whitegrid')
sns.set_style('whitegrid', {'grid.linestyle': '--',
                            'grid.color': '.7'})
```

### distplot - distribution plot

The distplot shows the distribution of a variable

Here the [distplot documentation](https://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=True, axlabel='Total Bill')
sns.despine()
```

```python
plt.figure(figsize=(12, 4))
a1=plt.subplot(1,3,1)
a2=plt.subplot(1,3,2)
a3=plt.subplot(1,3,3)
plt.tight_layout()
sns.distplot(ds['total_bill'], bins=25, color='g', hist=False, kde=True, axlabel='Total Bill', ax=a1)
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=False, axlabel='Total Bill', ax=a2)
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=True, axlabel='Total Bill', ax=a3)
```
