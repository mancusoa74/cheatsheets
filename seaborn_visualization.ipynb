# Seaborn Visualization Cheat Sheet

[Seaborn](https://seaborn.pydata.org/) is a Python data visualization library based on matplotlib. It works very weel with Pandas DataFrame

This cheat sheet, is a collection of the most commont type of plot that can be useful during the Explanatory Data Analysis **EDA** phase



## Common Import

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline
```

## Dataset load for testing

```python
ds = sns.load_dataset('tips')
```

```python
ds.head()
```

Show the available style with `sns.axes_style()`


```python
# set the styles with

sns.set_style('whitegrid')
sns.set_style('whitegrid', {'grid.linestyle': '--', 'grid.color': '.7'})
```

# Distribution Plot

Display distribution of variables

Main plots are:

- **distplot**
- **jointplot**
    
    

## distplot

The distplot shows the distribution of a variable

Here the [distplot documentation](https://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=True, axlabel='Total Bill')
sns.despine()

# set hist|kdw_kws dictionary with specific visialization params
```

```python
# variations 
plt.figure(figsize=(12, 4))
a1=plt.subplot(1,3,1)
a2=plt.subplot(1,3,2)
a3=plt.subplot(1,3,3)
plt.tight_layout()
sns.distplot(ds['total_bill'], bins=25, color='g', hist=False, kde=True, axlabel='Total Bill', ax=a1)
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=False, axlabel='Total Bill', ax=a2)
sns.distplot(ds['total_bill'], bins=25, color='g', hist=True, kde=True, axlabel='Total Bill', ax=a3)
sns.despine()
```

## jointplot

The jointplot shows the distribution and correlation of two variables

Here the [distplot documentation](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

There are 5 *kind* of joint plot

1. scatter
2. reg
3. kde
4. hex
5. resid


### scatter

Shows a scatter plot of the two variables.

```python
sns.jointplot(x='total_bill', y='tip', data=ds) #scatter kind is the default
```

### reg

Show a scatter plot of the two variable and a linear regression line

```python
sns.jointplot(x='total_bill', y='tip', data=ds, kind='reg', height=8, color='g')
```

### kde

```python
sns.jointplot(x='total_bill', y='tip', data=ds, kind='kde', height=8, color='red')
```

### hex

```python
sns.jointplot(x='total_bill', y='tip', data=ds, kind='hex', height=10, ratio=12, color='purple')
```

## pairplot

Shows a scatter plot between each combination of numeric variables in the dataset


```python
sns.pairplot(ds,  hue='smoker', palette='Set1', height=3, kind='reg')
```

# Categorical Plot

Plot categorical data

Main plots are:

- **catplot**
- **barplot**
- **countplot**
- **boxplot**
- **violinplot**
- **stripplot**
- **swarmplot**



## catplot

A generalization of categorical plot. Accept kind parameter


```python
sns.catplot(x='sex', y='total_bill', data=ds, kind='violin')
```

## barplot

Aggregates data by a category and apply some function. Default is **mean**


Here the [distplot documentation](https://seaborn.pydata.org/generated/seaborn.distplot.html)

```python
sns.barplot(x='sex', y='total_bill', data=ds, palette='Set1', estimator=np.mean, hue='smoker') #mean is the default
```

## countplot

Count occorrences of a variable. Same as barplot but the estimator is count


```python
sns.countplot(x='smoker', data=ds, palette='coolwarm', hue='sex')
```

## boxplot

Shows the distribution of categorical data

```python
plt.figure(figsize=(10,10))
sns.boxplot(x='day', y='total_bill', data=ds, width=0.5, palette='rainbow', hue='sex')
```

## violinplot

Show the density distribution of categorical data

```python
plt.figure(figsize=(10,6))
sns.violinplot(x='day', y='total_bill', data=ds, width=0.5, palette='Set1', hue='sex', split=True)
```

## striplot

Shows a scatterplot for a categorical variable

```python
plt.figure(figsize=(10,8))
sns.stripplot(x="day", y="total_bill", data=ds,jitter=False,hue='sex',palette='Set1',dodge=True)
```

## swarmplot

Similar to striplot but better distribute point on categorical axis to avoid overlapping

```python
plt.figure(figsize=(10,8))
sns.swarmplot(x="day", y="total_bill", data=ds, hue='sex',palette='Set1',dodge=True)
```
